```python
import pandas as pd
import os
```

```python
files = [file for file in os.listdir() if (file.startswith('COVID'))][2:]
files
```

```python
HealthDF = pd.read_csv('COVID_HealthIns.csv', index_col = 0)

for file in files:
    current_data = pd.read_csv(file, index_col = 0)
    HealthDF = pd.merge(HealthDF, current_data, how = 'left', on = ['State'])
    
HealthDF.head()
```

```python
list(HealthDF.columns)
```

```python
dropCols = [col for col in HealthDF.columns if ('cases' in col) or ('deaths' in col)]
HealthDF = HealthDF.drop(dropCols, axis = 1)
list(HealthDF.columns)
```

```python
HealthDF.head()
```

```python
COVID = pd.read_csv('COVID.csv', index_col = 0)
COVID.head()
```

```python
HealthDF = pd.merge(HealthDF, COVID, how = 'left', on = 'State')
HealthDF.head()
```

```python
HealthDF.to_csv('Healthcare.csv')
```
