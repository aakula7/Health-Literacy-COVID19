```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
Tab21 = pd.read_excel('Residence_all_tables.xlsx', 'Table 21 Population')
Tab24 = pd.read_excel('Residence_all_tables.xlsx', 'Table 24 Medicare Enrollment')
Tab27 = pd.read_excel('Residence_all_tables.xlsx', 'Table 27 Medicaid Enrollment')
Tab30 = pd.read_excel('Residence_all_tables.xlsx', 'Table 30 Private Enrollment')
```

```python
Tab21 = Tab21.rename(columns = {'Table 21: US Population Estimates by State - 1991 - 2014 (in Thousands)':'State',
                             'Unnamed: 24':'Population Per State (2014)'})
Tab21 = Tab21[['State', 'Population Per State (2014)']]
Tab21.head()
```

```python
Tab24 = Tab24.rename(columns = {'Table 24: Medicare Enrollment- 1991-2014 (in Thousands)':'State',
                             'Unnamed: 24':'Medicare Enrollment Per State (2014)'})
Tab24 = Tab24[['State', 'Medicare Enrollment Per State (2014)']]
Tab24.head()
```

```python
Tab27 = Tab27.rename(columns = {'Table 27: Medicaid Enrollment, 1991-2014 (in Thousands)':'State',
                             'Unnamed: 24':'Medicaid Enrollment Per State (2014)'})
Tab27 = Tab27[['State', 'Medicaid Enrollment Per State (2014)']]
Tab27.head()
```

```python
Tab30 = Tab30.rename(columns = {'Table 30: Private Health Insurance Enrollment -  2001-2014  (in Thousands)':'State',
                             'Unnamed: 14':'Private Enrollment Per State (2014)'})
Tab30 = Tab30[['State', 'Private Enrollment Per State (2014)']]
Tab30.head()
```

```python
HealthInsDF = [Tab21, Tab24, Tab27, Tab30]
HealthInsFinal = reduce(lambda left, right: pd.merge(left, right, on = 'State', how = 'outer'), HealthInsDF)
HealthInsFinal.head()
```

```python
COVIDCases = pd.read_csv('us-states.csv')
COVIDCases = COVIDCases.groupby(['state']).sum()
COVIDCases = COVIDCases.reset_index()
COVIDCases.drop([11, 36, 41, 49], inplace = True)
COVIDCases = COVIDCases.reset_index(drop = True)
COVIDCases.state
```

```python
COVIDCases = COVIDCases.rename(columns = {'state':'State'})
COVIDCases.head()
```

```python
COVIDHealthIns = pd.merge(COVIDCases, HealthInsFinal, how = 'left', on = 'State')
COVIDHealthIns
```

```python
COVIDHealthIns = COVIDHealthIns.drop(['fips'], axis = 1)
COVIDHealthIns
```
