```python
PerStateCapitaDF = [Tab11, Tab12, Tab13, Tab14, Tab16, Tab18, Tab19, Tab20]
PerStateCapitaFinal = reduce(lambda left, right: pd.merge(left, right, on = 'State', how = 'outer'), PerStateCapitaDF)
PerStateCapitaFinal.head()
```

```python
COVIDCases = pd.read_csv('us-states.csv')
COVIDCases = COVIDCases.groupby(['state']).sum()
COVIDCases = COVIDCases.reset_index()
COVIDCases.drop([11, 36, 41, 49], inplace = True)
COVIDCases = COVIDCases.reset_index(drop = True)
COVIDCases.state
```

```python
COVIDCases = COVIDCases.rename(columns = {'state':'State'})
COVIDCases.head()
```

```python
COVIDHealthPerStateCapita = pd.merge(COVIDCases, PerStateCapitaFinal, how = 'left', on = 'State')
COVIDHealthPerStateCapita
```

```python
COVIDHealthPerStateCapita = COVIDHealthPerStateCapita.drop(['fips'], axis = 1)
COVIDHealthPerStateCapita
```
