# Import Libraries

```python
import pandas as pd
import numpy as np
import os
```

```python
files = [file for file in os.listdir() if (file.startswith('COVID') & ('AllData' not in file) & ('AGI' not in file))][1:]

COVID_DF = pd.read_csv('COVID_AGI.csv', index_col = 0)

for file in files:
    current_data = pd.read_csv(file, index_col = 0)
    COVID_DF = pd.merge(COVID_DF, current_data, how = 'outer', on = ['State', 'cases', 'deaths'])

COVID_DF.head()
```

```python
StatePremDF = pd.read_excel('State Premiums.xlsx')
StatePremDF.head()
```

```python
COVID_DF = pd.merge(COVID_DF, StatePremDF, how = 'outer', on = ['State'])

COVID_DF.head()
```

```python
list(COVID_DF.columns)
```

```python
COVID_DF = COVID_DF.drop(columns = ['Public School Expenditure 2016(Thousands)_y'])
COVID_DF = COVID_DF.drop(columns = ['Physical Scientists 2018 (Percent)'])
COVID_DF = COVID_DF.rename(columns = {'Public School Expenditure 2016(Thousands)_x':'Public School Expenditure 2016(Thousands)'})
list(COVID_DF.columns)
```
