```python
import pandas as pd
import os
```

```python
files = files = [file for file in os.listdir() if file.startswith('COVID')][2:]
files
```

```python
IncomeDF = pd.read_csv('COVIDAGI.csv', index_col = 0)

for file in files:
    current_data = pd.read_csv(file, index_col = 0)
    IncomeDF = pd.merge(IncomeDF, current_data, how = 'outer', on = ['State', 'cases', 'deaths'])

IncomeDF.head()
```

```python
IncomeDF['All Ages SAIPE Poverty Universe'] = IncomeDF['All Ages SAIPE Poverty Universe'].str.replace(',', '').astype('int64')
IncomeDF['All Ages in Poverty Count'] = IncomeDF['All Ages in Poverty Count'].str.replace(',', '').astype('int64')
IncomeDF = IncomeDF.rename(columns = {'Civilian_labor_force_2018':'Civilian Labor Force',
                                     'Employed_2018':'Employed',
                                     'Unemployed_2018':'Unemployed',
                                     'Unemployment_rate_2018':'Unemployment Rate',
                                     'Median_Household_Income_2018':'Median Household Income'})

IncomeDF.head()
```

```python
IncomeDF = IncomeDF.drop(columns = ['cases', 'deaths', 'AGI Total', 'Unemployment Rate'])

IncomeDF.head()
```

```python
COVID = pd.read_csv('COVID.csv', index_col = 0)
COVID.head()
```

```python
IncomeDF = pd.merge(IncomeDF, COVID, how = 'left', on = 'State')
IncomeDF.head()
```

```python
IncomeDF.to_csv('IncomeMerged.csv')
```
