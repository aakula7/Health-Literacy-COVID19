# In area_name, we have states of U.S also the counties, we dont want counties, so we are removing the counties

```python
df2 = pd.DataFrame(columns=df.columns)

cond = df['Rural_urban_continuum_code_2013'].isnull()
rows = df.loc[cond, :]
df2 = df2.append(rows, ignore_index=True)
```

# Removing Unwanted Columns

```python
df = df2.drop(['Rural_urban_continuum_code_2013','Urban_influence_code_2013','Metro_2013'], axis =1)

```

# Replacing the NA Values as mean value of the column

```python
# fill missing values with mean column values
df.fillna(df.mean(), inplace=True)
```

# Join the COVID cases and Unemployment, so rename the columns as State

# Read the COVIS US States dataset

```python
covid_us_states = pd.read_csv('C:\\NachuMohan\\Saint_Peters_University\\GA_Summer2020\\us-states.csv')
```

# Removing unwanted columns

```python
covid_us_states = covid_us_states.groupby(['state']).sum()
covid_us_states = covid_us_states.reset_index()
covid_us_states.drop([11, 36, 41, 49], inplace = True)
covid_us_states = covid_us_states.reset_index(drop = True)
covid_us_states.head()
```

# Replace the column value  as State

```python
covid_us_states = covid_us_states.rename(columns = {'state':'State'})
covid_us_states.head()covid_us_states = covid_us_states.drop(['fips'], axis = 1)
covid_us_states
```

```python
covid_us_states = covid_us_states.drop(['fips'], axis = 1)
covid_us_states
```

# Merging the data Unemployment and COVID us states

```python
COVIDUnemployment = pd.merge(covid_us_states, df, how = 'left', on = 'State')
COVIDUnemployment
```
