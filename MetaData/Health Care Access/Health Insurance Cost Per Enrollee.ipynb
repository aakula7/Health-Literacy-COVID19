```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
Tab23 = pd.read_excel('Residence_all_tables.xlsx', 'Table 23 Medicare Per Enrollee')
Tab26 = pd.read_excel('Residence_all_tables.xlsx', 'Table 26 Medicaid Per Enrollee')
Tab29 = pd.read_excel('Residence_all_tables.xlsx', 'Table 29 Private Per Enrollee')
```

```python
Tab23 = Tab23.rename(columns = {
    'Table 23: Medicare Per Enrollee State Estimates by State of Residence (1991-2014) - Personal Health Care (Dollars)':'State',
                             'Unnamed: 14':'Medicare Health Insurance Per Enrollee (2014)'
})
Tab23 = Tab23[['State', 'Medicare Health Insurance Per Enrollee (2014)']]
Tab23.head()
```

```python
Tab26 = Tab26.rename(columns = {
    'Table 26: Medicaid Per Enrollee State Estimates by State of Residence (1991-2014) - Personal Health Care (Dollars)':'State',
                             'Unnamed: 14':'Medicaid Health Insurance Per Enrollee (2014)'
})
Tab26 = Tab26[['State', 'Medicaid Health Insurance Per Enrollee (2014)']]
Tab26.head()
```

```python
Tab29 = Tab29.rename(columns = {
    'Table 29: Private Health Insurance Per Enrollee State Estimates by State of Residence (2001-2014) - Personal Health Care (Dollars)':'State',
                             'Unnamed: 14':'Private Health Insurance Per Enrollee (2014)'
})
Tab29 = Tab29[['State', 'Private Health Insurance Per Enrollee (2014)']]
Tab29.head()
```

```python
InsPerEnrDF = [Tab23, Tab26, Tab29]
InsPerEnrFinal = reduce(lambda left, right: pd.merge(left, right, on = 'State', how = 'outer'), InsPerEnrDF)
InsPerEnrFinal.head()
```

```python
COVIDCases = pd.read_csv('COVID.csv', index_col = 0)
COVIDCases.head()
```

```python
COVIDCases = COVIDCases.rename(columns = {'state':'State'})
COVIDCases.head()
```

```python
COVIDInsPerEnr = pd.merge(COVIDCases, InsPerEnrFinal, how = 'left', on = 'State')
COVIDInsPerEnr.head()
```

```python
COVIDInsPerEnr.to_csv('COVID_InsPerEnr.csv')
```
