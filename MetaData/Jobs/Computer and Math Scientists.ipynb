```python
CompMath = CompMath.rename(columns = {'Table S-35':'State',
                                     'Unnamed: 16':'Computer Mathematical Scientists 2018',
                                     'Unnamed: 33':'All Occupations 2018',
                                     'Unnamed: 50':'Computer Mathematical Scientists 2018 (Percent)'})
CompMath.drop(range(0,4), inplace = True)
CompMath.drop(range(55,64), inplace = True)
CompMath = CompMath.reset_index(drop = True)
CompMath
```

```python
COVIDCases = pd.read_csv('us-states.csv')
COVIDCases = COVIDCases.groupby(['state']).sum()
COVIDCases = COVIDCases.reset_index()
COVIDCases.drop([11, 36, 41, 49], inplace = True)
COVIDCases = COVIDCases.reset_index(drop = True)
COVIDCases.state
```

```python
COVIDCases = COVIDCases.rename(columns = {'state':'State'})
COVIDCases.head()
```

```python
COVIDCompMath = pd.merge(COVIDCases, CompMath, how = 'left', on = 'State')
COVIDCompMath
```

```python
COVIDCompMath = COVIDCompMath.drop(['fips'], axis = 1)
COVIDCompMath
```
