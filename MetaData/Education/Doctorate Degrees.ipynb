```python
import pandas as pd
import numpy as np
```

```python
DocDeg = pd.read_csv('SED_Export_2020-05-25T16_55_23.182Z.csv')
DocDeg
```

```python
DocDeg = DocDeg[DocDeg['<Year>'] == 'Total']
DocDeg = DocDeg.reset_index(drop = True)
DocDeg
```

```python
DocDeg = DocDeg.drop(['Unnamed: 0', '<Year>'], axis = 1)
DocDeg
```

```python
DocDeg.drop(DocDeg.iloc[:, 3:], axis = 1, inplace = True)
DocDeg
```

```python
DocDeg = DocDeg.rename(columns = {'Unnamed: 1':'State',
                                 '2018':'Doctorate From Doc Inst 2018',
                                 '2018.1':'Doctorate From Bach Inst 2018'})
DocDeg
```

```python
DocDeg.drop(range(0,2), inplace = True)
DocDeg.drop(113, inplace = True)
DocDeg = DocDeg.reset_index(drop = True)
DocDeg
```

```python
DocDeg = DocDeg.groupby('State').sum()
DocDeg = DocDeg.reset_index()
DocDeg
```

```python
DocDeg.dtypes
```

```python
COVIDCases = pd.read_csv('us-states.csv')
COVIDCases = COVIDCases.groupby(['state']).sum()
COVIDCases = COVIDCases.reset_index()
COVIDCases.drop([11, 36, 41, 49], inplace = True)
COVIDCases = COVIDCases.reset_index(drop = True)
COVIDCases.state
```

```python
COVIDCases = COVIDCases.rename(columns = {'state':'State'})
COVIDCases
```

```python
COVIDCases.dtypes
```

```python
COVIDDocDeg = COVIDCases.merge(DocDeg, how = 'left', on = 'State')
COVIDDocDeg
```

```python
COVIDDocDeg['Doctorate From Doc Inst 2018'] = COVIDDocDeg['Doctorate From Doc Inst 2018'].str.replace(',', '')
COVIDDocDeg['Doctorate From Bach Inst 2018'] = COVIDDocDeg['Doctorate From Bach Inst 2018'].str.replace(',', '')
COVIDDocDeg
```

```python
COVIDDocDeg = COVIDDocDeg.astype({'Doctorate From Doc Inst 2018':'int64',
                                 'Doctorate From Bach Inst 2018':'int64'})
COVIDDocDeg
```

```python
COVIDDocDeg['Total Doctorate Degrees 2018'] = COVIDDocDeg['Doctorate From Doc Inst 2018'] + COVIDDocDeg['Doctorate From Bach Inst 2018']
COVIDDocDeg
```

```python
COVIDDocDeg.to_csv('COVID_Doc.csv')
```
