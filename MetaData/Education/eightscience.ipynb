```python
import pandas as pd
import numpy as np
```

```python
eightscience = pd.read_excel('eighth-grade-science-performance.xlsx')
eightscience.head()
```

```python
eightscience.info()
```

```python
eightscience = eightscience.rename(columns = {'Table S-7' : 'State',
                                   'Unnamed: 1' : '8th 2009 science',
                                   'Unnamed: 2' : '8th 2011 science',
                                   'Unnamed: 3' : '8th 2015 science'})

eightscience.drop(range(0,4), inplace = True)

eightscience.head()
```

```python
eightscience = eightscience.reset_index(drop = True)

eightscience.head()
```

```python
eightscience.isna().sum()
```

```python
eightscience.State
```

```python
eightscience.drop(range(51,60), inplace = True)
eightscience.State
```

```python
eightscience.isna().sum()
```

```python
eightscience = eightscience.fillna(eightscience.mean())
eightscience
```

```python
eightscience = eightscience.astype({'8th 2009 science':'int32',
                                    '8th 2011 science':'int32',
                                    '8th 2015 science':'int32'})

eightscience
```

```python
COVIDCases = pd.read_csv('us-states.csv')
COVIDCases
```

```python
COVIDCases = COVIDCases.groupby(['state']).sum()
COVIDCases = COVIDCases.reset_index()
COVIDCases
```

```python
COVIDCases.state
```

```python
COVID8science = pd.merge(COVIDCases, eightscience, how = 'left', on = 'State')
COVID8science
```
