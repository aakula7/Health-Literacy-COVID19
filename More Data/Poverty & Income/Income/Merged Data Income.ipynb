```python
import pandas as pd
import numpy as np
```

```python
IncomeDF = pd.read_excel('county5.xls')
IncomeDF.head()
```

```python
IncomeDF.info()
```

```python
IncomeDF = IncomeDF.drop(columns = [c for c in IncomeDF.columns if c not in ['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 10',
                                                                             'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13',
                                                                             'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16']])

IncomeDF.head()
```

```python
IncomeDF = IncomeDF.rename(columns = {'Unnamed: 1':'County Name',
                                      'Unnamed: 2':'County Code',
                                     'Unnamed: 10':'County Income Mean',
                                     'Unnamed: 11':'First Quintile',
                                     'Unnamed: 12':'Second Quintile',
                                     'Unnamed: 13':'Third Quintile',
                                     'Unnamed: 14':'Fourth Quintile',
                                     'Unnamed: 15':'Fifth Quintile',
                                     'Unnamed: 16':'95% Mean'})

IncomeDF.head()
```

```python
IncomeDF = IncomeDF.drop([0 , 1])
IncomeDF = IncomeDF.astype({'County Code':'int64'})
IncomeDF = IncomeDF.reset_index(drop = True)
IncomeDF.head()
```

```python
IncomeDF = IncomeDF.dropna()
IncomeDF.isnull().sum()
```

```python
COVID = pd.read_csv('COVID_Case.csv')
COVID.head()
```

```python
COVID = COVID.drop(columns = [c for c in COVID.columns if c not in ['countyFIPS', 'County Name', '8/18/20']])
COVID = COVID.rename(columns = {'countyFIPS':'County Code',
                                'County Name':'County',
                               '8/18/20':'Cases'})
COVID.head()
```

```python
COVID.isnull().sum()
```

```python
IncomeCOVID = pd.merge(IncomeDF, COVID, how = 'left', on = 'County Code')
IncomeCOVID.head()
```

```python
IncomeCOVID = IncomeCOVID.drop(columns = ['County Name'])
IncomeCOVID.head()
```

```python
IncomeCOVID = IncomeCOVID[['County Code', 'County', 'Cases', 'County Income Mean', 'First Quintile', 'Second Quintile',
                          'Third Quintile', 'Fourth Quintile', 'Fifth Quintile', '95% Mean']]
IncomeCOVID.head()
```

```python
IncomeCOVID.isnull().sum()
```

```python
IncomeCovid = IncomeCOVID.dropna()
IncomeCovid.isnull().sum()
```

```python
IncomeCOVID.to_csv('Income.csv')
```
