```python
import pandas as pd
import numpy as np
```

```python
MDI_DF = pd.read_excel('county-level-mdi-rates-2017.xls')
MDI_DF.head()
```

```python
MDI_DF = MDI_DF.drop(columns = ['Standard Error'])
MDI_DF = MDI_DF.rename(columns = {'county':'County Code'})
MDI_DF = MDI_DF.dropna()
MDI_DF = MDI_DF.astype({'County Code':'int64'})
MDI_DF.head()
```

```python
MDI_DF.isnull().sum()
```

```python
COVID = pd.read_csv('COVID_Case.csv')
COVID.head()
```

```python
COVID = COVID.drop(columns = [c for c in COVID.columns if c not in ['countyFIPS', 'County Name', '8/18/20']])
COVID = COVID.rename(columns = {'countyFIPS':'County Code',
                                'County Name':'County',
                               '8/18/20':'Cases'})
COVID.head()
```

```python
COVID.isnull().sum()
```

```python
MDI_COVID = pd.merge(MDI_DF, COVID, how = 'left', on = 'County Code')
MDI_COVID.head()
```

```python
MDI_COVID = MDI_COVID[['County Code', 'County', 'MDI rate', 'Cases']]
MDI_COVID.head()
```

```python
MDI_COVID.to_csv('MDI.csv')
```
