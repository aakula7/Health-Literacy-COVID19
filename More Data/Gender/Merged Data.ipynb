```python
import pandas as pd
import numpy as np
```

```python
DemDF = pd.read_csv('Demographics.csv', encoding = 'ISO-8859-1')
DemDF.head()
```

```python
DemDF = DemDF.drop(columns = ['SUMLEV', 'STATE', 'COUNTY', 'STNAME', 'YEAR', 'AGEGRP'])
DemDF = DemDF.drop(columns = [c for c in DemDF.columns if c not in ['CTYNAME', 'TOT_POP', 'TOT_MALE', 'TOT_FEMALE']])
DemDF = DemDF.groupby('CTYNAME').sum()
DemDF = DemDF.reset_index()
DemDF.head()
```

```python
COVID = pd.read_csv('COVID_Case.csv')
COVID.head()
```

```python
COVID = COVID.drop(columns = [c for c in COVID.columns if c not in ['County Name', '8/18/20']])
COVID = COVID.rename(columns = {'County Name':'CTYNAME',
                               '8/18/20':'Cases'})
COVID.head()
```

```python
Dem_COVID = pd.merge(DemDF, COVID, how = 'left', on = 'CTYNAME')
Dem_COVID.head()
```

```python
Dem_COVID = Dem_COVID.dropna()
Dem_COVID.isnull().sum()
```

```python
Dem_COVID.to_csv('GenderDem.csv')
```
