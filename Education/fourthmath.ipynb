```python
fourmath = fourmath.rename(columns = {'Table S-1' : 'State',
                                   'Unnamed: 1' : '4th 2000 Math',
                                   'Unnamed: 2' : '4th 2003 Math',
                                   'Unnamed: 3' : '4th 2005 Math',
                                   'Unnamed: 4' : '4th 2007 Math',
                                   'Unnamed: 5' : '4th 2009 Math',
                                   'Unnamed: 6' : '4th 2011 Math',
                                   'Unnamed: 7' : '4th 2013 Math',
                                   'Unnamed: 8' : '4th 2015 Math',
                                   'Unnamed: 9' : '4th 2017 Math',
                                   'Unnamed: 10': '4th 2019 Math' })

fourmath.drop(range(0,4), inplace = True)

fourmath.head()
```

```python
fourmath = fourmath.reset_index(drop = True)

fourmath.head()
```

```python
fourmath.isna().sum()
```
