```python
import pandas as pd
import numpy as np
```

```python
ELL = pd.read_excel('ELL Students.xls')
ELL
```

```python
ELL = ELL.drop(['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9',
                'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13','Unnamed: 14', 'Unnamed: 15',
                'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20'], axis = 1)

ELL.head()
```

```python
ELL = ELL.rename(columns = {'Table 204.20. English language learner (ELL) students enrolled in public elementary and secondary schools, by state: Selected years, fall 2000 \n              through fall 2017':'State',
                           'Unnamed: 10': 'ELL Students Enrolled 2017'})

ELL.head()
```

```python
ELL.drop(range(0,4), inplace = True)
ELL.drop(range(64,69), inplace = True)
```

```python
ELL.tail(15)
```

```python
ELL = ELL.dropna()
ELL.isnull().sum()
```

```python
ELL = ELL.reset_index(drop = True)
ELL
```

```python
COVIDCases = pd.read_csv('us-states.csv')
COVIDCases = COVIDCases.groupby(['state']).sum()
COVIDCases = COVIDCases.reset_index()
COVIDCases.drop([11, 36, 41, 49], inplace = True)
COVIDCases = COVIDCases.reset_index(drop = True)
COVIDCases.state
```

```python
COVIDELL = pd.concat([COVIDCases, ELL], axis = 1)
COVIDELL.head()
```

```python
COVIDELL = COVIDELL.drop(['State'], axis = 1)
COVIDELL = COVIDELL.rename(columns = {'state':'State'})
COVIDELL
```

```python
COVIDELL = COVIDELL.astype({'ELL Students Enrolled 2017':'int32'})
COVIDELL
```

```python
COVIDELL = COVIDELL.drop(['fips'], axis = 1)
COVIDELL
```

```python
COVIDELL.to_csv('COVID_ELL.csv')
```
