```python
import pandas as pd
import os
```

```python
files = [file for file in os.listdir() if file.startswith('COVID')][2:]
files
```

```python
EducatDF = pd.read_csv('COVID_Education_Merged.csv', index_col = 0)

for file in files:
    current_data = pd.read_csv(file, index_col = 0)
    EducatDF = pd.merge(EducatDF, current_data, how = 'outer', on = ['State', 'cases', 'deaths'])

EducatDF.head()
```

```python
list(EducatDF.columns)
```

```python
EducatDF = EducatDF.drop(columns = ['Public School Expenditure 2016(Thousands)_y', 'cases', 
                                    'deaths', 'Doctorate From Doc Inst 2018', 'Doctorate From Bach Inst 2018'])
EducatDF = EducatDF.drop(columns = [col for col in EducatDF.columns if 'Percent' in col])
EducatDF = EducatDF.rename(columns = {'Public School Expenditure 2016(Thousands)_x':
                                      'Public School Expenditure 2016(Thousands)'})

list(EducatDF.columns)
```

```python
EducatDF.head()
```

```python
Cases = pd.read_csv('COVID.csv', index_col = 0)
Cases.head()
```

```python
EducatDF = pd.merge(EducatDF, Cases, how = 'left', on = 'State')
EducatDF.head()
```

```python
EducatDF.to_csv('Educat.csv')
```
