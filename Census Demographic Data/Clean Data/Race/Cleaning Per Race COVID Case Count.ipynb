```python
%matplotlib inline

import pandas as pd
import numpy as np
from scipy.stats import spearmanr
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pylab import rcParams
from matplotlib.backends.backend_agg import FigureCanvasAgg

plt.style.use('seaborn-darkgrid')
```

```python
RaceCovid = pd.read_csv('Race Data Entry.csv')
RaceCovid.head()
```

```python
RaceCovid.info()
```

```python
RaceCovid = RaceCovid[RaceCovid['Date'] == 20200712]
RaceCovid.head()
```

```python
columnsKeep = ['State']

for col in list(RaceCovid.columns):
    if 'Cases_' in col:
        columnsKeep.append(col)
        
RaceCovid = RaceCovid[columnsKeep]
RaceCovid.head()
```

```python
RaceCovid = RaceCovid.drop(columns = ['Cases_LatinX', 'Cases_Multiracial', 'Cases_Other', 'Cases_Unknown', 
                                      'Cases_Ethnicity_NonHispanic', 'Cases_Ethnicity_Unknown'])
RaceCovid.head()
```

```python
RaceCovid = RaceCovid.fillna(0)
RaceCovid.head()
```

```python
RaceCovid = RaceCovid.rename(columns = {'State':'Code',
                                        'Cases_Total':'cases',
                                       'Cases_White':'White Cases',
                                       'Cases_Black':'Black or African American Cases',
                                       'Cases_Asian':'Asian Cases',
                                       'Cases_AIAN':'American Indian and Alaska Native Cases',
                                       'Cases_NHPI':'Native Hawaiian and Other Pacific Islander Cases',
                                       'Cases_Ethnicity_Hispanic':'Hispanic or Latino Cases'})

RaceCovid = RaceCovid.astype({'White Cases':'int64','Black or African American Cases':'int64', 'Asian Cases':'int64',
                              'American Indian and Alaska Native Cases':'int64',
                              'Native Hawaiian and Other Pacific Islander Cases':'int64',
                              'Hispanic or Latino Cases':'int64'})

RaceCovid.head()
```

```python
StatesDF = pd.read_csv('csvData.csv')
StatesDF.head()
```

```python
RaceCovid = pd.merge(RaceCovid, StatesDF, how = 'left', on = 'Code')
RaceCovid.head()
```

```python
RaceCovid = RaceCovid[['State', 'cases', 'White Cases', 'Black or African American Cases', 'Asian Cases',
                       'American Indian and Alaska Native Cases', 'Native Hawaiian and Other Pacific Islander Cases',
                       'Hispanic or Latino Cases']]

RaceCovid = RaceCovid.dropna()

RaceCovid.head()
```

```python
df = pd.read_csv('COVIDdemographic.csv')
df.head()
```

```python
df = df.rename(columns = {'Total population_HISPANIC OR LATINO':'Hispanic or Latino'})
df.head()
```

```python
df['Hispanic or Latino'] = (df['Total Population'])-(df['White'] + df['Black or African American'] + 
                                                     df['American Indian and Alaska Native'] + df['Asian'] + 
                                                     df['Native Hawaiian and Other Pacific Islander'])

df.head()
```

```python
df = df[['State', 'cases', 'deaths', 'Total Population', 'White', 'Black or African American', 
         'American Indian and Alaska Native', 'Asian', 'Native Hawaiian and Other Pacific Islander', 
         'Hispanic or Latino']]

df.head()
```

```python
df = df.drop(columns = ['cases', 'deaths'])
df.head()
```

```python
RaceCovid = pd.merge(RaceCovid, df, how = 'left', on = 'State')
RaceCovid.head()
```

```python
RaceCovid = RaceCovid.rename(columns = {'White':'White Population',
                                       'Black or African American':'Black or African American Population',
                                       'Asian':'Asian Population',
                                       'American Indian and Alaska Native':'American Indian and Alaska Native Population',
                                       'Native Hawaiian and Other Pacific Islander':'Native Hawaiian and Other Pacific Islander Population',
                                       'Hispanic or Latino':'Hispanic or Latino Population'})

RaceCovid.head()
```

```python
RaceCovid.info()
```

```python
RaceCate = ['White', 'Black or African American', 'Asian', 'American Indian and Alaska Native', 
            'Native Hawaiian and Other Pacific Islander', 'Hispanic or Latino']

for Race in RaceCate:
    RaceCovid['{} Cases Per 100K'.format(Race)] = (RaceCovid['{} Cases'.format(Race)]/RaceCovid['{} Population'.format(Race)]) * 100000
    
RaceCovid.head()
```

```python
RaceCovid.to_csv('RacialCovidCount.csv')
```

```python
RaceCovidPer100K = RaceCovid

columnsToKeep = ['State', 'cases']

for col in list(RaceCovidPer100K.columns):
    if 'Per 100K' in col:
        columnsToKeep.append(col)
        
RaceCovidPer100K = RaceCovidPer100K[columnsToKeep]
RaceCovidPer100K.head()
```

```python
RaceCovidPer100K.to_csv('RaceCovidPer100K.csv')
```
