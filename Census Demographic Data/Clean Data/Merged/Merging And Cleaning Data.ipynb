```python
import pandas as pd
import os
```

```python
DemographDF = pd.read_csv('COVIDdemographic.csv')
DemographDF.head()
```

```python
DemographDF = DemographDF.drop(columns = ['cases', 'deaths', 'RACE_Total population'])
DemographDF = DemographDF.rename(columns = {'Total population_HISPANIC OR LATINO':'Hispanic or Latino'})
DemographDF.head()
```

```python
DemographDF['Hispanic or Latino'] = (DemographDF['Total Population'])-(DemographDF['White'] + DemographDF['Black or African American'] + 
                                                     DemographDF['American Indian and Alaska Native'] + DemographDF['Asian'] + 
                                                     DemographDF['Native Hawaiian and Other Pacific Islander'])

DemographDF.head()
```

```python
DemographDF = DemographDF.drop(columns = ['Total Population'])
DemographDF.head()
```

```python
COVID = pd.read_csv('COVID.csv', index_col = 0)
COVID.head()
```

```python
DemographDF = pd.merge(DemographDF, COVID, how = 'left', on = 'State')
DemographDF.head()
```

```python
DemographDF.to_csv('DemographicsMerged.csv')
```
