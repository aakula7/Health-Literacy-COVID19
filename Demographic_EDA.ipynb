# Instead of NA values, we could see there is a value N

Replacing the N values as 0

```python
df1 = df.replace(to_replace = 'N', value = 0) 
```

First N value found in state West Virginia, so instead of taking whole dataframe, checking for the state West Virginia whether the N value has replaced with value 0

```python
df2 = pd.DataFrame(columns=df.columns)

cond = df1['Geographic Area Name'] == 'West Virginia'
rows = df1.loc[cond, :]
df2 = df2.append(rows, ignore_index=True)
```
